```haskell
-- add_three::int->int->int->int --Int
add_three:: Int->Int->Int->Int --参数之间由 -> 分隔，而与回传值之间并无特殊差异。回传值是最后一项，参数就是前三项
add_three x y z = x + y + z
add_three 1 2 3
```

```haskell
add_three:: Int, Int, Int->Int
```

```haskell
factorial :: Integer -> Integer --Int 有界，integer可以非常大
factorial n = product [1..n]
factorial 120
```
